### Statistical Data Graphics
#### In these examples we are going to use a library called Seaborn

```python
import numpy as np
import numpy.random as randn
import pandas as pd

from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5,4
```

```python
import seaborn as sb
sb.set_style('whitegrid')
```

```python
cars = pd.read_csv("../Data/mtcars.csv")
cars.head()
print(type(cars))
```

#### Checking distributions using Histograms

```python
cars.index = cars.model
mpg = cars['mpg']
mpg.plot(kind='hist')
```

#### Using Seaborn

Note here that we are using the `displot` function - not `distplot` as this is being deprecated. 

```python
sb.displot(mpg)
```

#### Scatterplots

```python
cars.plot(kind='scatter', x='hp', y='mpg', c=['darkgray'], s=150)
```

#### Using Seaborn

```python
sb.regplot(x='hp', y='mpg', data=cars, scatter=True)
```

#### Generate a scatterplot matrix

```python
sb.pairplot(cars)
```

That's a lot of data so lets create a _subset_ ....

```python
cars_subset = cars[['mpg', 'disp', 'hp', 'wt']]
sb.pairplot(cars_subset)
```

#### Boxplots

```python
cars.boxplot(column='mpg', by='am')
cars.boxplot(column='wt', by='am')
```

#### Using Seaborn ....

```python
sb.boxplot(x='am', y='mpg', data=cars, palette='hls')
```
