## Plot Building

```python
import numpy as np
import numpy.random as randn
import pandas as pd

from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
```

### Line Charts
#### Create a line chart from a `list` object

```python
x = range(1,10)
y = [1,2,3,4,0,4,2,3,1]

plt.plot(x,y)
```

#### Plot a line chart from a Pandas object

```python
cars = pd.read_csv("../Data/mtcars.csv")
cars.head()
```

```python
mpg = cars["mpg"]
mpg.plot()
```

To plot a number of variables (as lines) just select the columns as appropriate.

```python
df1 = cars[['cyl', 'wt', 'mpg']]
df1.plot()
```

### Bar charts
#### Create a bar chart from a list

```python
plt.bar(x,y)
```

#### Create a bar chart from a Pandas object

```python
mpg.plot(kind="bar")
```

```python
mpg.plot(kind="barh")
```

### Create a pie chart

```python
x = [1,2,3,4,0.5]
plt.pie(x)
plt.show()
```

It is possible to save a plot as an image

```python
plt.pie(x)
plt.savefig("../Data/pie_chart.png")
plt.show()
```
