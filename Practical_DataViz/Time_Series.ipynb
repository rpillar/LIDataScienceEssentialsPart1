### Visualizing Time Series

```python
import numpy as np
import numpy.random as randn
import pandas as pd

from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5,4
```

#### The simplest Time Series plot

```python
sales = pd.read_csv("../Data/Superstore-Sales.csv", index_col='Order Date', encoding='cp1252', parse_dates=True)
sales.head()
```

```python
sales['Order Quantity'].plot()
```

```python
sales_sample = sales.sample(n=100, random_state=25, axis=0)
plt.xlabel('Order Date')
plt.ylabel('Order Quantity')
plt.title('SuperStore Sales')

sales_sample['Order Quantity'].plot()
```
