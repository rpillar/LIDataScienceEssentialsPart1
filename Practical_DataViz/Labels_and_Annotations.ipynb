## Data Viz - Labels and Annotaions

```python
import numpy as np
import numpy.random as randn
import pandas as pd

from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 8,4
```

### Labeling plot features
#### Using the Functional method

```python
x = range(1,10)
y = [1,2,3,4,.5,4,3,2,1]
plt.bar(x,y)
plt.xlabel('x-axis label')
plt.ylabel('y-axis label')
```

#### The object-oriented method

```python
cars = pd.read_csv("../Data/mtcars.csv")
cars.head()
```

```python
mpg = cars.mpg

fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])

mpg.plot()

ax.set_xticks(range(32))
ax.set_xticklabels(cars.model, rotation=60, fontsize='medium')
ax.set_title('MPG - MTCARS Dataset')
ax.set_xlabel('model')
ax.set_ylabel('MPG')

```

#### Add a legend to your plot

There (as always) is two different ways to do this - here is the object oriented method ....

```python
fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])

mpg.plot()

ax.set_xticks(range(32))
ax.set_xticklabels(cars.model, rotation=60, fontsize='medium')
ax.set_title('MPG - MTCARS Dataset')
ax.set_xlabel('model')
ax.set_ylabel('MPG')

ax.legend(loc='best')
```

#### Finally - we can anotate a plot

```python
fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])

mpg.plot()

ax.set_xticks(range(32))
ax.set_xticklabels(cars.model, rotation=60, fontsize='medium')
ax.set_title('MPG - MTCARS Dataset')
ax.set_xlabel('model')
ax.set_ylabel('MPG')

ax.legend(loc='best')

ax.set_ylim([0, 45])
ax.annotate('Toyota Corolla', xy=(19, 33.9), xytext=(21, 35), arrowprops=dict(facecolor='black', shrink=0.05))
```
