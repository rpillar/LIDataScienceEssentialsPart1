### Defining Elements of a Plot
Plot elements add context to a plot and thereby help convey its meaning to viewers

In these examples we will be an `object oriented` approach to building plots using the following steps:-
* Create a blank figure object
* Add axes to the figure
* Generate plot(s) within the figure object
* Specify plotting and layout parametersfor the plots within the figure.

```python
import numpy as np
import numpy.random as randn
import pandas as pd

from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
```

To set some parameters for showing the plots within our notebooks :-

```python
%matplotlib inline
rcParams['figure.figsize'] = 5,4
```

#### Defining axes, ticks and grids

```python
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]

fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])
ax.plot(x,y)
```

We can set `limits` to our axes :-

```python
fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])

ax.set_xlim([1, 9])
ax.set_ylim([-1, 5])

ax.set_xticks([0,1,2,4,5,6,8,9,10])
ax.set_yticks([0,1,2,3,4,5])

# add a grid
ax.grid()
ax.plot(x,y)
```

#### Generating multiple plots in a single figure with subplots

```python
fig = plt.figure()
fig, (ax1, ax2) = plt.subplots(1,2)

ax1.plot(x)
ax2.plot(x,y)
```
