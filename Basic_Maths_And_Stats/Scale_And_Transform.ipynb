## Scale and Transform your data
Its important to do this because :-
- So that differing magnitude between variables do not produce erroneous or misleaading statistics
- To prepare your data for machine learning

There are two ways to scale your data :-
- Normalization - putting each observation on a realtive scale between the values of 0 and 1
- Standardization - rescaling data so that it has a zero mean and unit variance

You can use the `scikit-learn` library to :-
- scale your data
- center your data
- normalize your data
- bin your data
- impute your data

```python
import numpy as np
import pandas as pd
import scipy

import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sb

import sklearn
from sklearn import preprocessing
from sklearn.preprocessing import scale
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5,4
sb.set_style('whitegrid')
```

```python
cars = pd.read_csv("../Data/mtcars.csv")
cars.head()
```

### Normalizing and transforming features with MinMaxScaler()  and fit_transform()

```python
mpg = cars.mpg
plt.plot(mpg)
```

```python
mpg.describe()
```

One of the first things we need to do is ensure that our `mpg` variable is in a one-column matrix. we can use the `reshape` function to do this

```python
mpg_matrix = mpg.values.reshape(-1, 1)
```

```python
scaled = preprocessing.MinMaxScaler()
scaled_mpg = scaled.fit_transform(mpg_matrix)
plt.plot(scaled_mpg)
```

As we can see the data is scaled - in terms of valaues  between `0` and `1`. These are the defaults - but these defaults can be changed if required.

```python
scaled = preprocessing.MinMaxScaler(feature_range=(0,10))
scaled_mpg = scaled.fit_transform(mpg_matrix)
plt.plot(scaled_mpg)
```

### Standardising variables

The data we are using in these examples is fairly simple - but particularly when taking data from a number of different sources (where that data may have different scales) it is important that all the data has the same scale in order to avoid any bias in outcomes.

Standardization is used on the data values that are normally distributed. Further, by applying standardization, we tend to make the mean of the dataset as 0 and the standard deviation equivalent to 1.

```python
standard_mpg = scale(mpg, axis=0, with_mean=False, with_std=False)
plt.plot(standard_mpg)
```

If we were to use the `scale()` function without any options - only defaults then the plot looks different (the shape is the same but the axis values are different). The `mean` has been centered to `zero` and the distribution how has `unit variance`.

```python
standard_mpg = scale(mpg)
plt.plot(standard_mpg)
```
