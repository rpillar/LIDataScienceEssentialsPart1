## Parametric Correlation Analysis

This is a method that can be used to find correlations between linearly related continuous numeric variables.
The Pearson Correlation is measurec by the coefficient 'R' where :-
- R = 1 indicates a strong positive relationship
- R = 0 indicates no correlation
- R = -1 indicates a strong negative relationship

The assumptions here are that :-
- your data is normally distributed
- you have continuous numeric variables
- your variables are linearly related

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sb
from pylab import rcParams

import scipy
from scipy.stats import pearsonr
```

```python
# %matplotlib inline
rcParams['figure.figsize'] = 8,4
plt.style.use('seaborn-whitegrid')
```

### The Pearson Correlation

```python
cars = pd.read_csv("../Data/mtcars.csv")
cars.head()
```

```python
sub_df = cars[['mpg', 'hp', 'qsec', 'wt']]
sb.pairplot(sub_df)
```

### Use scipy to calculate the Pearson correlation coefficient

```python
mpg = cars['mpg']
hp = cars['hp']
qsec = cars['qsec']
wt = cars['wt']

pearsonr_coeffient, p_value = pearsonr(mpg, hp)
print(f"PearsonR Correlation Coefficient : {pearsonr_coeffient:.3f}")
```

```python
pearsonr_coeffient, p_value = pearsonr(mpg, qsec)
print(f"PearsonR Correlation Coefficient : {pearsonr_coeffient:.3f}")
```

```python
pearsonr_coeffient, p_value = pearsonr(mpg, wt)
print(f"PearsonR Correlation Coefficient : {pearsonr_coeffient:.3f}")
```

### Using pandas to calculate the Pearson correlation coeffcient

```python
corr = sub_df.corr()
corr
```

### Using Seaborn to visualize the Pearson correlation coefficient

```python
sb.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns.values)
```
